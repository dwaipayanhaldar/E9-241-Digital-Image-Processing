```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = np.ones((64, 64), dtype=np.uint8)
img[16:48, 16:48] = 0
img[24:40, 24:40] = 1

plt.figure(figsize=(12, 6))
plt.imshow(img, cmap='gray')
plt.title('Original Image')
plt.axis('off')
```

```python
#majority filter
def majority_filter(image, kernel_size=3):
    pad = kernel_size // 2
    padded_image = np.pad(image, pad, mode='constant', constant_values=1)
    filtered_image = np.zeros_like(image)

    for i in range(image.shape[0]):
        for j in range(image.shape[1]):
            region = padded_image[i:i+kernel_size, j:j+kernel_size]
            ones_count = np.sum(region)
            if ones_count > (kernel_size * kernel_size) / 2:
                filtered_image[i, j] = 1
            else:
                filtered_image[i, j] = 0

    return filtered_image
```

```python
img_filtered = majority_filter(img, kernel_size=3)
```

```python
plt.figure(figsize=(12, 6))
plt.imshow(img_filtered, cmap='gray')
plt.title('Filtered Image with Majority Filter')
plt.axis('off')
```

```python
output_1 = np.logical_or(img_filtered, img)  # To keep the last image in a variable for future use
```

```python
plt.figure(figsize=(12, 6))
plt.imshow(output_1, cmap='gray')
plt.title('Filtered Image with Majority Filter')
plt.axis('off')
```
